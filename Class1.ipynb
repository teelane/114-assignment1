    Set largest to first value in the list

    for each number in the list starting from the second item:
        if the number is greater than largest:
            update largest of this number

```python
nums =[12, 4, 27, 3, 188, 50, 7]

def find_largest(nums):
    if not nums:
        return "No values to compare"

    largest = nums[0]

    for n in nums[1:]:
        if n > largest:
            largest = n

    return largest
print(find_largest(nums))    
```
