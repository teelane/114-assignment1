1 # Queues
2
3 A que is an ordered collection of elements where items are added and removed from *opposite* ends, commonly referred to as the front and back end or rear of the queue. The ordering principle is referred to as FIFO.

```python
stack = [] 

stack.append("A")
stack.append("B")
stack.append("C")

if stack:  # Check if stack is not empty
    Top = stack.pop()
else:
    Top = None

# Print the entire stack and the value that was removed
print("Stack after first pop:", stack)
print("Value removed (Top):", Top)

# Pop the next value from the stack and print it
if stack:
    next_value = stack.pop()
    print("Next value removed:", next_value)
else:
    print("Stack is empty, nothing to pop.")

# Push the value 'D' onto the stack
stack.append("D")

# Pop again and print the removed value
if stack:
    removed_value = stack.pop()
    print("Value removed after pushing D:", removed_value)
else:
    print("Stack is empty, nothing to pop.")

# Print the entire stack
print("Final stack:", stack)
```
