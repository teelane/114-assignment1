```python
#Challenge 1: define the class stack

class Stack:
    def __init__(self):
        self.items = []

#challenge 2: define the pop method

    def is_empty(self):
        return len(self.items) == 0
    
#challenge 3: peek() method

    def push(self, item):
        self.items.append(item)

#challenge 4

    def pop(self):
        
        if self.is_empty():  # Check if stack is empty
            return None
        return self.items.pop()  # Removes and returns last item

#challenge 5

    def peek(self):
        if self.is_empty():
            return None
        return self.items[-1]

    def size(self):
        return len(self.items)

```

```python
s = Stack()

s.push(5)
s.push(10)
s.push(15)

print("Stack peek: ", s.peek())
print("Stack pop: ", s.pop())
print("Stack size: ", s.size())
```

```python
#Challenge 6 create a class named Queue with a constructor that initializes an empty list.

class Queue:
    def __init__(self):
        self.items = []

#challenge 7: add an is_empty method just like before

    def is_empty(self):
        return len(self.items) == 0
    
#challenge 8: write the enqueue() method to add items to the back of the queue using append()

    def enqueue(self, item):
        self.items.append(item)

#challenge 9:create que method then check if que is empty

    def dequeue(self):
        if self.is_empty():
            return None
        return self.items.pop(0)
    
#challenge 10 

    def size(self):
        return len(self.items)
    
    
```

```python
#testing the queue
q = Queue()

q.enqueue("A")
q.enqueue("B")
q.enqueue("C")

print("Queue dequeue: ", q.dequeue())
print("Queue size: ", q.size())
```
