```python
#====================================
# Single List Implementation
#====================================

class Node:
    def __init__ (self, value):
        self.value=value
        self.next=None

class LinkedList():
    def __init__(self):
        self.head=None

    def append(self, value):
        # Add a new node at the end of the list
        new_node = Node(value)
        
        # if the list is empty, the new node becomes the head
        if self.head is None:
            self.head = new_node
            return
        
        # Otherwise, walk to the end and attach it.
        current = self.head
        while current.next is not None:
            current = current.next
        current.next = new_node

    # Exercise 1: print_list, Print all values in a list.
    def print_list(self):
        current=self.head

        if current is None:
            print("List is empty")
            return
        while current is not None:
            print(current.value)
            current = current.next

    # Exercise 3: Count the Nodes in a list
    def count_nodes(self):
        count = 0
        current = self.head
        while current is not None:
            count += 1
            current = current.next
        return count
    
    # Exercise 4: Insert at the beginning
    def prepend(self, value):
        new_node = Node(value)
        new_node.next = self.head
        self.head = new_node

    # Exercise 5: Remove a Node With a Specific Value
    def remove(self, value):
        if self.head is None:
            return
        
        if self.head.value == value:
            self.head = self.head.next
            return
        
        current = self.head
        while current.next is not None:
            if current.next.value == value:
                current.next = current.next.next
                return
            current = current.next 

# Exercise 2: Create List and Append Values and Print the Result.

myList = LinkedList()
myList.append(10)
myList.append(20)
myList.append(30)

myList.print_list()

# Challenge 3 result.
print("Node Count:", myList.count_nodes())

# Challenge 4 Result:
myList.prepend(5)
print("List (After Append)") 
myList.print_list()

# Challenge 5 Result:
myList.remove(10)
print("List (After Remove)")
myList.print_list()




```
